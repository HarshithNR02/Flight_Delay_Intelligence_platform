```python
import pandas as pd
import os

dfs = []
folder = 'dataset/airports'
for f in sorted(os.listdir(folder)):
    if f.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder, f))
        dfs.append(df)
        print(f'{f}: {len(df):,} rows')

t100 = pd.concat(dfs, ignore_index=True)
print(f'\nTotal: {len(t100):,} rows')
print(f'Columns: {t100.columns.tolist()}')
print(t100.head(3).to_string())
```

```python
t100.to_parquet('dataset/airports/t100_clean.parquet', index=False)
print(f"Saved: {len(t100):,} rows")
print(f"Airports (origin): {t100['ORIGIN'].nunique()}")
print(f"Year range: {t100['YEAR'].min()} - {t100['YEAR'].max()}")
print(f"Months per year:")
print(t100.groupby('YEAR')['MONTH'].nunique())
```
