```python
for year in ['2023', '2024', '2025']:
    path = f"dataset/{year}"
    if os.path.exists(path):
        files = sorted(os.listdir(path))
        print(f"\n{year}: {len(files)} files")
        for f in files[:3]:
            print(f"  {f}")
        if len(files) > 3:
            print(f"  ... ({len(files)} total)")
```

## Load and Concatenate All CSV Files

```python
dfs = []
for year in ['2023', '2024', '2025']:
    folder = Path(f"dataset/{year}")
    for f in sorted(folder.glob("*.csv")):
        df = pd.read_csv(f, low_memory=False)
        dfs.append(df)
        print(f"Loaded {f.name}: {len(df):,} rows | {df['YEAR'].iloc[0]}-{df['MONTH'].iloc[0]:02d}")

flights = pd.concat(dfs, ignore_index=True)
print(f"\nTotal: {flights.shape[0]:,} rows x {flights.shape[1]} columns")

print("\nRows per year-month:")
print(flights.groupby(['YEAR', 'MONTH']).size().to_string())

flights['FL_DATE'] = pd.to_datetime(flights['FL_DATE'])

flights.to_parquet("dataset/flights_raw.parquet", index=False)
print(f"\nSaved to dataset/flights_raw.parquet ({os.path.getsize('dataset/flights_raw.parquet') / 1e9:.2f} GB)")
```
